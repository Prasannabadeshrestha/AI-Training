```python
labels = data[["play"]]
print('labels here: \n')
display(labels)

print('**'*50)

features=data.drop(columns=['play'])
print('features here: ')
display(features)

print('**'*50)

display(data[['temp', 'humidity']])
```

```python
#to see unique data

unique_outlook=list(data['outlook'].unique())
display(unique_outlook)

#count how many sunny outlook is there in data

display(data['outlook'].loc[data['outlook']=='sunny'].count())

sunny_count=data['outlook'].loc[data['outlook']=='sunny'].count()
print(f'total sunny: {sunny_count}')

#getting count

count_var={}
for val in unique_outlook:
    count_var[val]=data['outlook'].loc[data['outlook']==val].count()

    print(count_var)

try:
    plt.bar(count_var.keys(),count_var.values())
except:
    pass
```

```python
#to see unique data

unique_temp=list(data['temp'].unique())
display(unique_temp)

#count how many sunny outlook is there in data

display(data['temp'].loc[data['temp']=='hot'].count())

hot_count=data['temp'].loc[data['temp']=='hot'].count()
print(f'total hot: {sunny_count}')

#getting count

count_var={}
for val in unique_temp:
    count_var[val]=data['temp'].loc[data['temp']==val].count()

    print(count_var)

try:
    plt.bar(count_var.keys(),count_var.values())
except:
    pass
```

Convert all no to 0 and yes for 1 for easy processing in labels

```python
def convert_to_number(x):
    if x=='no':
        return 0
    elif x=='yes':
        return 1

print('Before Change: ')
display(labels)
print('**'*25)

labels['play']=labels['play'].apply(convert_to_number)
print('After Change: ')
display(labels)
print('**'*25)
```
